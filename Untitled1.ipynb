```python
from bokeh.io import show
from bokeh.plotting import figure
```

```python
from bokeh.layouts import gridplot
```

```python
from bokeh.io import output_notebook
output_notebook()
```

```python
graph_line = figure(
    plot_height=400, 
    plot_width=900, 
    title='PRESTAMOS', 
    tools='' 
)
```

```python
# Definimos los  datos que usara nuestra grafica
graph_line.line(
    x=datos.Anios, 
    y=datos.Incidencia, 
    color='#f4d341', 
    legend='Frecuencia', 
    line_width=5
)
graph_line.line(x=datos.Anios, y=datos.Brecha, color='#42f4bf', legend='Brecha',line_width=5)
graph_line.line(x=datos.Anios, y=datos.Severidad, color='#f44171', legend='Severidad',line_width=5)
```
